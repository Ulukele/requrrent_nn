```python
# Gain Ratio
import math

def get_H(data, tf):
    cnt = data[tf].count()
    groups_cnts = [ x / cnt for x in list(data[tf].value_counts()) if x != 0]
    return -sum( x * math.log2(x) for x in groups_cnts)

def get_Ha(data, feature_name, category, tf):
    selection = data[data[feature_name] == category]
    ratio = data[data[feature_name] == category][feature_name].count() / data[feature_name].count()
    return get_H(selection, tf) * ratio


def get_SI(data, feature_name, category):
    ratio = data[data[feature_name] == category][feature_name].count() / data[feature_name].count()
    if ratio == 0:
        return 0
    return ratio * math.log2(ratio)


def gain_ratio(data, target_feature):
    # Information Gain
    H = get_H(data, target_feature)
    IG = {}
    for feature_name in data.columns:
        feature = data[feature_name]
        IG[feature_name] = H - sum([
            get_Ha(data, feature_name, category, target_feature)
            for category in feature.unique()
        ])

    # Split Info
    SI = {}
    for feature_name in data.columns:
        feature = data[feature_name]
        SI[feature_name] = -sum([
            get_SI(data, feature_name, category)
            for category in feature.unique()
        ])
    return { key: IG[key] / SI[key] for key in IG.keys()}
```

```python
def colors_from_values(values, palette_name):
    normalized = (values - min(values)) / (max(values) - min(values))
    indices = np.nan_to_num(np.round(normalized * (len(values) - 1))).astype(np.int32)
    palette = sns.color_palette(palette_name, len(values))
    return np.array(palette).take(indices, axis=0)

def gain_ratio_info(data, target_features, target_feature):
    gr = gain_ratio(data, target_feature)
    gr = dict(sorted(gr.items(), key=lambda item: item[1]))
    keys = [k for k in gr.keys() if k not in target_features]
    values = [gr[k] for k in keys]
    ax = sns.barplot(x=values, y=keys, palette=colors_from_values(values, "YlOrRd"))
    ax.bar_label(ax.containers[0])
    return ax
```

```python
from sklearn.preprocessing import StandardScaler
scaler_x = StandardScaler()
scaler_y = StandardScaler()
x = scaler_x.fit_transform(x)
y = scaler_y.fit_transform(y)
x_t = scaler_x.transform(x_t)
y_t = scaler_y.transform(y_t)
```

```python
from Model import Model
act = lambda x: np.tanh(x)
der = lambda x: 1 - x**2
model = Model(x.shape[1], act, der, 1337)
model.add_layer(13, 1)
```

```python
lf = lambda x, y: np.mean((x - y)**2)
losses = model.fit(x, y, x_t, y_t, lf, sl=4, a=0.0001, epochs=40)
```

```python
import matplotlib.pyplot as plt
def draw(errs_, t_errs_):
    sns.set(rc = {'figure.figsize':(15,8)})
    plt.plot(np.array(range(errs_.size)), errs_, color='r', label='train')
    plt.plot(np.array(range(errs_.size)), t_errs_, color='g', label='test')
    plt.xlabel("epoch")
    plt.ylabel("loss")
    plt.legend()
    plt.show()
```

```python
errs = np.array([l[0] for l in losses])
t_errs = np.array([l[1] for l in losses])
draw(errs, t_errs)
```
